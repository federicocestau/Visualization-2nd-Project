##IDJ

```python
idj_now_url = 'https://www.idjnow.com/chauvet-dj-gigbar-move-system.html'
```

```python
# if website gives a '403' error code use headers to fix error 
headers = {'User-Agent': 'Chrome/39.0.2171.95'}
idj_now_response = requests.get(idj_now_url, headers=headers)
idj_now_soup = BeautifulSoup(idj_now_response.text, 'html.parser') 
# scrape the website for the logo 
idj_logo = idj_now_soup.find_all('a', class_='logo') 
# use for loop to grab the logo of the website
for logo in idj_logo: 
    idj_logo = logo.img['src']
idj_descr = idj_now_soup.find('div', class_='tabs-content overview open').get_text()
IDJP = idj_now_soup.find_all("span",class_="price")
IDJPFinal = IDJP[1].getText()
```

```python
IDJData = pd.DataFrame(
    {"Brand Name": "IDJ","Logo":[idj_logo],
     "Description":idj_descr,
    "Price":IDJPFinal,"Link":idj_now_url})
IDJData
```
